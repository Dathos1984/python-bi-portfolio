# Matplotlib

```python
import matplotlib.pyplot  as plt
import numpy as np
```

```python
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
```

```python
pupulation = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]
```

```python
plt.plot(year, pupulation, marker='o', color='red', linestyle='solid', linewidth=2, markersize=5)
```

```python
plt.scatter(year, pupulation, color='blue', marker='o', s=100, alpha=0.5)
```

```python
plt.bar(year, pupulation, color='green', alpha=0.5, width=5,edgecolor='black')
```

```python
plt.barh(  year, pupulation, color='green', alpha=0.5, height=5,edgecolor='black')
```

```python
values = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8]
```

```python
plt.hist(values, bins=3, color='purple', alpha=0.7, edgecolor='black')
```

```python
x= np.random.normal(250,10,500)
```

# Marker line Color

```python
x = np.array(range(0,10))
y = np.array(range(0,10))
```

```python
plt.plot(x,y,'o-.b')
```

```python
plt.plot(x,y,'o-.b')
plt.title('Population Growth Over Years')
plt.ylabel('Population (in billions)')
plt.xlabel('Year')
plt.grid(True)
plt.xticks(list(range(0, 11)))
plt.yticks(list(range(0, 11)))
```
