# Join DataFrames

```python
a = pd.DataFrame(
    {"name": ["Jhon", "Alex", "Felipe"],
     "age" :[30, 31, 32]}
)
```

```python
b = pd.DataFrame(
    {"name": ["Jhon", "Alex", "Felipe"],
     "eventos" :[323, 234, 245]}
)
```

```python
pd.merge(a, b)
```
